```
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

url = "https://pmb.upi.edu/download"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers, verify=False)
soup = BeautifulSoup(response.text, "html.parser")

pdf_links = []

# Ambil semua link yang mengandung PDF
for a in soup.find_all("a", href=True):
    href = a["href"]
    if ".pdf" in href.lower():
        full_url = urljoin(url, href)
        pdf_links.append(full_url)

print(f"Total PDF ditemukan: {len(pdf_links)}")
```

```
for link in tqdm(pdf_links):
    try:
        filename = link.split("/")[-1]
        filepath = os.path.join(save_dir, filename)

        res = requests.get(link, headers=headers)

        if res.status_code == 200:
            with open(filepath, "wb") as f:
                f.write(res.content)
        else:
            print(f"Gagal download: {link}")

    except Exception as e:
        print(f"Error: {e}")
```

```
!pip install pdfplumber
import pandas as pd

import pdfplumber

def is_scanned_pdf(filepath):
    try:
        with pdfplumber.open(filepath) as pdf:
            for page in pdf.pages[:3]:  # cek 3 halaman awal
                text = page.extract_text()
                if text and len(text.strip()) > 50:
                    return False  # text PDF
        return True  # kemungkinan scan
    except:
        return True
```

```
import pytesseract
from pdf2image import convert_from_path

def ocr_pdf(filepath):
    images = convert_from_path(filepath)
    text_all = ""

    for img in images:
        text = pytesseract.image_to_string(img, lang='ind+eng')
        text_all += text + "\n"

    return text_all
```

```
def extract_text_pdf(filepath):
    if is_scanned_pdf(filepath):
        return ocr_pdf(filepath), "scanned"
    else:
        with pdfplumber.open(filepath) as pdf:
            text = ""
            for page in pdf.pages:
                text += page.extract_text() or ""
        return text, "text"
```

```
import re

def generate_filename(text, idx):
    text = text.lower()

    # kategori
    if "registrasi" in text:
        category = "registrasi"
    elif "panduan" in text:
        category = "panduan"
    elif "beasiswa" in text:
        category = "beasiswa"
    else:
        category = "dokumen"

    # tahun
    year = re.findall(r'20\d{2}', text)
    year = year[0] if year else "unknown"

    return f"upi_{category}_{year}_{idx}.pdf"
```

```
import os

folder = "/content/drive/MyDrive/Dataset_PMB_UPI"

results = []

for i, file in enumerate(os.listdir(folder)):
    if file.endswith(".pdf"):
        path = os.path.join(folder, file)

        text, tipe = extract_text_pdf(path)
        new_name = generate_filename(text, i)

        new_path = os.path.join(folder, new_name)

        os.rename(path, new_path)

        results.append({
            "old_name": file,
            "new_name": new_name,
            "type": tipe
        })

print("Selesai!")
```
